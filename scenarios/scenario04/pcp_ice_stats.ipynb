```python


def get_year (dat, year):
    ind = np.where(np.floor(dat[:,0]*1e-4) == year)
    return dat[ind,:]

def get_month (dat, month):
    year = np.floor(dat[:,0]*1e-4)
    ind = np.where(np.floor(dat[:,0]*1e-2-year*100) == month)
    return dat[ind,:]

years = np.arange(2003, 2016)
res_m = np.zeros(len(years))
res_mts = np.zeros(len(years))

dat_y = get_year(my_dat, 2003)
print (my_dat.shape)
print (dat_y.shape)
dat_2 = dat_y[0,:,1]
dat_25 = get_month(dat_y[0], 5)
print (dat_25.shape)
print (np.where(dat_25[0,:,3]) == 1)
dat_251 = dat_25[:,np.where(dat_25[0,:,3]== 1),:]
print (dat_251.shape)
plt.plot(dat_251[0,:,0]-20030500)
plt.figure()
plt.plot(dat_25[0,:,-1])

for i in range(len(years)):
    dat_y = get_year(my_dat, years[i])
    dat_2 = dat_y[0,:,1]
    res_m[i] = np.mean(dat_2[np.where(np.isnan(dat_2) == False)])
    summer = np.zeros((5,31))
    #for m in range(5):
    #    mon = get_month(dat_y[0], m + 5)
    #    summer[m,:]= mon[:,:,1]
    res_mts = np.mean(summer[np.where(np.isnan(summer)==False)])

print (res_m)
print (res_mts)
plt.figure()
plt.plot(my_dat[:1000,3])
locs = np.where(np.isnan(my_dat[:,3]) == False)
print(locs)
```
